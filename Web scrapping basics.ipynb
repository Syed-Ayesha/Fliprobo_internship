# Problem 2

```python
page1 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating")
page1
```

```python
page1.content
```

```python
soup1 = BeautifulSoup(page1.content,'html.parser')
soup1
```

```python
movies = soup1.find_all('div',class_ = "lister-item mode-advanced")
movies[0]
```

```python
first_movie =movies[0]
```

```python
first_movie.h3.a.text
```

```python
first_movie.h3
```

```python
first_movie.find("span",class_="lister-item-year text-muted unbold").text
```

```python
first_movie.find('div',class_="inline-block ratings-imdb-rating" )['data-value']
```

```python
Name = []
Year = []
Rating = []
for i in movies:
    Name.append(i.h3.a.text)
    Year.append(i.find("span",class_="lister-item-year text-muted unbold").text)
    Rating.append(i.find('div',class_="inline-block ratings-imdb-rating" )['data-value'])
    
```

```python
data = list(zip(Name,Year,Rating))
```

```python
import pandas as pd
df = pd.DataFrame(data,columns=['Name','Year','Rating'])
```

```python
df.to_csv('file1.csv')
```

# Problem 4

```python
page3 = requests.get("https://bookpage.com/reviews")
page3
```

```python
page3.content
```

```python
soup3 = BeautifulSoup(page3.content,'html.parser')
soup3
```

```python
books = soup3.find_all('div',class_ = "bp-block article-info")
books[0]
```

```python
first_book =books[0]
```

```python
first_book.find('h4').a.text
```

```python
first_book.p.text.replace("\n",'')
```

```python
first_book.find('p', class_="genre-links hidden-phone").text.replace("\n"," ")
```

```python
first_book.find('p', class_="excerpt").text.replace("\n"," ")
```

```python
Book_Name = []
Author_Name = []
Genre=[]
Book_Review=[]

for t in books[0:5]:
    Book_Name.append(t.find('h4').a.text)
    Author_Name.append(t.p.text.replace("\n",''))
    Genre.append(t.find('p', class_="genre-links hidden-phone").text.replace("\n"," "))
    Book_Review.append(t.find('p', class_="excerpt").text.replace("\n"," "))
```

```python
data2 = list(zip(Book_Name,Author_Name,Genre,Book_Review))
```

```python
import pandas as pd
df2 = pd.DataFrame(data2,columns=['Book_Name','Author_Name','Genre','Book_Review'])
```

```python
df2
```
